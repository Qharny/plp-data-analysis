```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a simple DataFrame from scratch
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 500, 350, 100],
    'Stock': [15, 25, 30, 10, 50],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)
print("\n")

# Basic data inspection
print("DataFrame info:")
print(df.info())
print("\n")

print("Summary statistics:")
print(df.describe())
print("\n")

# Selecting and filtering data
print("Products costing more than $500:")
print(df[df['Price'] > 500])
print("\n")

print("Accessories only:")
print(df[df['Category'] == 'Accessories'])
print("\n")

# Sorting data
print("Products sorted by price (highest to lowest):")
print(df.sort_values('Price', ascending=False))
print("\n")

# Adding a new column
df['Value'] = df['Price'] * df['Stock']
print("DataFrame with total value column:")
print(df)
print("\n")

# Grouping and aggregating data
print("Average price by category:")
print(df.groupby('Category')['Price'].mean())
print("\n")

print("Total inventory value by category:")
print(df.groupby('Category')['Value'].sum())
```
